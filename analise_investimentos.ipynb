```python
abev = yf.download('ABEV3.SA', start='2010-01-01')['Close']
odont = yf.download('ODPV3.SA', start='2010-01-01')['Close']
vivo = yf.download('VIVT3.SA', start='2010-01-01')['Close']
petrobras = yf.download('PETR4.SA', start='2010-01-01')['Close']
bb = yf.download('BBAS3.SA', start='2010-01-01')['Close']
ibov = yf.download('^BVSP', start='2010-01-01')['Close']
```

```python
df_teste = df.copy()
for i in df_teste.columns[0:]:
  df_teste[i]= (df_teste[i] - df_teste[i].shift(1))/df_teste[i].shift(1)
```

```python
import plotly.express as px
import numpy as np

figura = px.line(title='Retorno Diário Percentual')

for i in df_teste.columns:

  figura.add_scatter(x=df_teste.index, y=df_teste[i], name=i, line={'width': 0.6})
  figura.update_yaxes(tickformat=".2%")  # Formata os rótulos do eixo y como porcentagem

figura.show()
```

```python
df_norm = df.copy()

for i in df_norm.columns[0:]:
  df_norm[i] = df_norm[i] / df_norm[i][0]
```
