```python
import yfinance as yf
```

```python
weg = yf.download("WEGE3.SA",start="2020-01-01")#br - .sa
msft = yf.download("MSFT", start="2020-01-01")
```

```python
msft.head()#ADJ CLOSE - fechamento do ativo(liquidez)
```

#tratando weg

```python
weg.head()
```

```python
weg_ohlc = weg[['Open', 'High', 'Low', 'Close']]
```

```python
weg_ohlc.head()
```

```python
weg_ohlc.plot();
```

```python
import pandas as pd
```

```python
weg_ohlc.to_csv("weg_ohlc.csv",sep=';', decimal='.')
```

```python
weg_exportado = pd.read_csv('weg_ohlc.csv',sep=';', decimal='.')
```

```python
weg_exportado.head()
```

```python
weg_exportado.plot()
```

```python
weg_exportado.index#--> date p/ index
```

```python
weg_exportado.dtypes
```

```python
weg_exportado.index = weg_exportado['Date']
```

```python
weg_exportado.head()
```

```python
type(weg_exportado.index)
```

```python
weg_exportado['Date'] = pd.to_datetime(weg_exportado['Date'])
```

```python
weg_exportado.drop('Date',axis = 1, inplace=True)
```

```python
weg_exportado.head()
```
