````python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib

import yfinance as yf
````

````python
acoes = ['GOLL4.SA', 'CVCB3.SA', 'WEGE3.SA', 'MGLU3.SA', 'TOTS3.SA', 'BOVA11.SA']
````

````python
acoes_df = pd.DataFrame()
for i in acoes:
  acoes_df[i]=yf.download(i, start='2010-01-01', end='2014-01-01')['Close']
````

````python
acoes_df = acoes_df.rename(columns ={'GOLL4.SA': 'GOL',	'CVCB3.SA': 'CVC', 	'WEGE3.SA':'WEGE',	'MGLU3.SA': 'MGLU', 	'TOTS3.SA':'TOTVS',	'BOVA11.SA':'BOV'})
````

````python
acoes_df['RS_GOL']= (acoes_df['GOL'] / acoes_df['GOL'].shift(1))-1
acoes_df['RS_CVC']= (acoes_df['CVC'] / acoes_df['CVC'].shift(1))-1
acoes_df['RS_WEGE']= (acoes_df['WEGE'] / acoes_df['WEGE'].shift(1))-1
acoes_df['RS_MGLU']= (acoes_df['MGLU'] / acoes_df['MGLU'].shift(1))-1
acoes_df['RS_TOTVS']= (acoes_df['TOTVS'] / acoes_df['TOTVS'].shift(1))-1
acoes_df['RS_BOV']= (acoes_df['BOV'] / acoes_df['BOV'].shift(1))-1
````
